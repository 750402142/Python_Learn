```python
#shift的使用 让数据向前或向后移动 正数表示向前移动 负数表示向后移动
sales = pd.DataFrame({"sales":[2000,2500,3000,4000,3500,4500]})
print(sales)
sales["next_month"] = sales.shift(1)
print(sales["next_month"])
sales["diff_value"] = sales["next_month"] - sales["sales"]
print(sales["diff_value"])
sales["add_percent"] = sales["diff_value"]/sales["next_month"]*100
```

```python
dt1 = pd.DataFrame({
    '工号': [101,102,103,104],
    '姓名': ['ALex','Alice','Alpha','Beta'],
    '部分ID':['pt-1','pt-2','pt-3','pt-5'],
    '业绩':[10000,8000,9000,5000]
})
dt1
```

```python
dt2 = pd.DataFrame({
    '工号': [104,105,106,107],
    '姓名': ['ALex-1','Alice-1','gamma','theta'],
    '部分ID':['pt-1','pt-2','pt-3','pt-x'],
    '是否在职':['是','否','是','是']
})
dt2
```

```python
pd.concat([dt1,dt2])#不加参数合并,只能找轴0方向合并,确实的数据采用Nan填充
pd.concat([dt1,dt2],axis=1)#指定横着合并
pd.concat([dt1,dt2],axis = 0,join = "inner")#join默认为取并集,设置inner取交集
pd.concat([dt1,dt2],axis = 0,join = "inner",ignore_index=True)#ignore_index重置索引
pd.concat([dt1,dt2],axis = 1,join = "inner",keys = ["表1","表2"])#ignore_index重置索引
```

```python
import json
import os
table_result = []
filenames = [i for i in os.listdir("./country_json") if i.endswith("json")]
for index,i in enumerate(filenames):
    with open(f"./country_json/{i}","r",encoding="utf-8") as f:
        data = json.loads(f.read())
    date = pd.DataFrame(data['data']['list'])['date'] # 提取出date这个列
    today = pd.DataFrame([i['today'] for i in data['data']['list']]) # 提取today这个列
    today.columns = [f"td_{i}" for i in today.columns] # 修改today的名称
    total = pd.DataFrame([i['total'] for i in data['data']['list']]) # 提取total这个列
    total.columns = [f"tt_{i}" for i in total.columns]# 修改total这个列名
    table = pd.concat([date,today,total],axis=1)
    table.insert(0,"ids",i.strip(".json"))
    table_result.append(table) # 将三个子表进行合并
all_tables = pd.concat(table_result)
all_tables.to_csv("./all_tables.csv",index=False)
```

```python
drinks= pd.read_csv("./drinks.csv")
#统计每个州国家的数量
drinks.groupby(by = ["continent"])["country"].count()
#统计每个州啤酒均值消耗最大的数量的索引
drinks.groupby(by = ["continent"])["beer_servings"].mean().idxmax()
#描述性统计
drinks.groupby(by = ["continent"])["wine_servings"].describe()
#
drinks.groupby(by = ["continent"])["beer_servings"].agg(["max","min","mean"])
drinks.groupby(by = ["continent"])["spirit_servings"].agg([lambda x:x.max()-x.min(),lambda x:x.quantile(0.75)-x.quantile(0.25)])
drinks.groupby(by = ["continent"]).agg({
    "spirit_servings":["min","mean"],
    "wine_servings":["max",lambda x:x.max() - x.min()]
})
drinks["drinks_beer"] = drinks.groupby(by = ["continent","country"])["beer_servings"].rank()

drinks["drinks_wine"] = drinks.groupby(by = ["continent","country"])["wine_servings"].rank()
drinks
```
