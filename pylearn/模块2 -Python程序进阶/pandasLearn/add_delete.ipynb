```python
data = pd.read_csv("./pandas_test.csv")
data.head(2)
#data.groupby(按谁分)[对谁做操作].做的操作上什么
#求每个学校的数学成绩最大值
data.groupby(by = "School")["math_scores"].max()
data.groupby(by = "School")["math_scores"].min()
data.groupby(by = ["School","Grade"])["math_scores"].mean()#mean用来求均值
```

```python
data.groupby(by = ["School","gender"])["Height"].max()
```

```python
#按学校和年级分组 统计身高的最大值 最小值 平均值
data.groupby(by = ["School","Grade"])["Height"].describe()[['min','max',"mean"]]
```

```python
#按学校和性别分组,统计四级成绩的平方差
data.groupby(by = ["School","Grade"])["cet4_scores"].std()
```

```python
#按性别学校和性别,统计数学成绩的极差
dt1 = data.groupby(by = ["School","gender"])["math_scores"].describe()[["max","min"]].reset_index()
print(dt1)
dt1["极差"] = dt1["max"] - dt1["min"]#在原有列的基础上再加一行
print(dt1)
```

```python
#按照学校和年级分组,统计身高的最大值,体重的均值
dt1 = data.groupby(by = ["School","Grade"])["Height","Weight"].describe()
dt1.loc[:,[("Height","max"),("Weight","mean")]].reset_index()
```

```python
#聚合函数
#按照学校和性别分组,统计身高的最大值,最小值和均值
data.groupby(by = ["School","Grade"]).agg({
    "Height":["max","min","mean"]
}).reset_index()
```

```python
#按照学校和年级分组,统计身高的最大值,最小值,体重的均值和标准差
data.groupby(by = ["School","Grade"]).agg({
    "Height":["max","min"],
    "Weight":["mean","std"]
}).reset_index()#reset_index()的用处是生成一个具体的表格
```

```python
data.groupby(by = ["School","Grade"])["age"].agg(lambda x:x.max())
```

```python
from itertools import count
#数学成绩超过平均值的人数 
def my_function(x):
    return sum(x>x.mean())/x.shape[0]
data.groupby(by = ["School","Grade"]).agg({
    "Name":["size"],
    "age":[("平均值","mean"),("众数",lambda x:x.value_counts().idxmax())],
    "Height":[("最大值","max"),("最小值","min"),("标准差","std")],
    "Weight":[("平均值","mean"),("极差",lambda x:x.max()-x.min()),("超过平均值的人数",lambda x:sum(x>x.mean()))],
    "math_scores":[("方差","var"),("标准差","std"),("超过平均值的人数占比",lambda x:sum(x>x.mean())/x.shape[0]),("成绩比例2",my_function)],
    "cet4_scores":[("四级成绩大于425的人数",lambda x:sum(x>425))]
}
    
).reset_index()
#数学成绩超过平均人数的占比
```
