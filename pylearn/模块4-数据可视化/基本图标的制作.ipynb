```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def get_round_rect(rects, ec='black', fc='#96c8d6'):
    bb = rects.get_bbox()
    patch = FancyBboxPatch((bb.xmin, bb.ymin),
                           abs(bb.width), abs(bb.height),
                           boxstyle="Round, pad=0, rounding_size=0.05",
                           ec=ec, fc=fc, linewidth=1,
                           mutation_aspect=4,
                           mutation_scale=1)
    return patch

# 假设这是你的数据
data = {
    'Country': ['Country A', 'Country B', 'Country C', 'Country D', 'Country E'],
    'Pensions': [100, 120, 90, 110, 80],
    'Income': [150, 130, 140, 160, 120]
}

x = np.arange(len(data['Country']))
colors = ['#464ca6', '#6472ea', '#7fc9da', '#eec568', '#df8342']

plt.figure(figsize=(14, 7), dpi=100)

# 创建柱状图
rects1 = plt.bar(x - 0.2, data['Pensions'], width=0.4)
rects2 = plt.bar(x, data['Income'], align='edge', width=0.4)

ax = plt.gca()

# 遍历每个柱子并应用圆角矩形补丁
for i, rect_pair in enumerate(zip(rects1, rects2)):
    rect1, rect2 = rect_pair
    rect1.remove()
    rect2.remove()
    patch1 = get_round_rect(rect1, fc=colors[i % len(colors)])
    patch2 = get_round_rect(rect2, fc='orange')
    ax.add_patch(patch1)
    ax.add_patch(patch2)

# 设置 x 轴标签为国家名称
plt.xticks(ticks=x, labels=data['Country'])

# 设置标题和标签
plt.title('Comparison of Pensions and Income by Country')
plt.xlabel('Country')
plt.ylabel('Amount')

# 添加网格线以增强可读性
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 添加图例
plt.legend(['Pensions', 'Income'])

plt.show()
```
