```python
x = np.arange(0.0,2.0,0.01)
y1 = np.sin(2*np.pi*x +1) -.2
y2 = np.cos(2*np.pi*x +1) -.2
plt.rcParams['font.sans-serif'] = ['SimHei']#设置全局字体为黑体来解决对中文不友好的问题
plt.rcParams['axes.unicode_minus'] = False#让负数可以标准显示
```

```python
# 设置点的个数
m = 20
n = 256 * m + 1  # 总点数
# 变量的取值范围
tx = np.linspace(0, 24 * np.pi, n)
# 初始化 x 和 y 数组
x = np.zeros(n)
y = np.zeros(n)
y1 = np.zeros(n)
y2 = np.zeros(n)
y3 = np.zeros(n)
# 计算函数的值
for i in range(n):
    t = tx[i]
    x[i] = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - (np.sin(t / 12))**5)
    y[i] = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - (np.sin(t / 12))**5)
    y1[i] = np.cos(t) * (np.exp(np.cos(t)) + 2 * np.cos(4 * t) - (np.sin(t / 12))**5)
    y2[i] = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) + (np.sin(t / 12))**5)
    y3[i] = np.cos(t) * (np.exp(np.cos(t)) + 2 * np.cos(4 * t) + (np.sin(t / 12))**5)
value_y = []
value_y.extend([y,y1,y2,y3])
colors = ['#b5d1e2','orange','pink','#3992ea']
```

```python

fig,axs = plt.subplots(
    nrows = 2,#子图的行数
    ncols = 2,#子图的列数
    figsize = (12,8),#画布的大小
    dpi = 100,#分辨率
    sharex=True,#共享x轴
    sharey = True,#共享y轴
    facecolor = "None",#设置背景色为透明
    frameon = True,#开启还不得边框
    linewidth = 4,#设置边框的线条宽度
    edgecolor = "cyan",#边框的颜色
    width_ratios=[2,2],#子图每列的宽度比
    height_ratios=[2,2],#子图每行的宽度比例
    subplot_kw = {'facecolor':'none'},#子图的背景色为透明
    gridspec_kw={'hspace':0, 'wspace':0},#子图的间距
)
for index,ax in enumerate(axs.flatten()):
    ax.plot(x,value_y[index],color = colors[index])

plt.show()
```
