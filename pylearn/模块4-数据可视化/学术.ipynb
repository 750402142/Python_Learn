```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sys
print(sys.version)
print(pd.__version__)
print(np.__version__)
print(matplotlib.__version__)
```

```python
x = np.arange(0.0,2.0,0.01)
y1 = np.sin(2*np.pi*x +1) -.2
y2 = np.cos(2*np.pi*x +1) -.2
plt.rcParams['font.sans-serif'] = ['SimHei']#设置全局字体为黑体来解决对中文不友好的问题
plt.rcParams['axes.unicode_minus'] = False#让负数可以标准显示
```

```python
plt.figure(figsize = (8,5),dpi = 100)#创建画布
plt.plot(x,y1,#第二步,绘制基本的图形
         color = 'red',#线条颜色
         linewidth = 1.2,#调整线条灯粗细
         label = 'y = sin(2*np.pi*x +1) -1',#该线条所对应的名称
         )
plt.plot(x,y2,#第二步,绘制基本的图形
         color = 'blue',#
         linewidth = 1.2,
         label = 'y = cos(2*np.pi*x +1) -1',
         )
plt.xlabel('这是测试的x轴名称')#4.1设置x轴坐标轴名称
plt.ylabel('这是测试的y轴名称')#4.2设置y轴名称
plt.title('This is figure title')#4.3图表的标题
plt.grid(True)#添加网格线
plt.xlim(0,2)#4.5调整x轴刻度的范围
plt.ylim(-1.25,1)#4.6调整y轴刻度的范围
plt.legend(loc= 2)#添加图表的图例,由于图表没有名称,因此需要回到第二部,添加图表的名称
#plt.savefig("./figure/second.png",dpi= 200)
plt.show()#3.展示图形
```

# 绘图主题的使用
- 内置主题
- 第三方主题(商业):matplotx
- 第三方主题(学术)

```python
plt.rcParams['font.sans-serif'] = ['Times New Roman']
with plt.style.context(['fast','seaborn-v0_8-colorblind']):
    plt.figure(figsize = (10,6),dpi = 100)#创建画布
    plt.plot(x,y1,label = 'y = sin(2*np.pi*x +1) -1',)
    plt.plot(x,y2,label = 'y = cos(2*np.pi*x +1) -1',)
    plt.xlabel('这是测试的x轴名称',fontdict={'family':'simhei'})#4.1设置x轴坐标轴名称
    plt.ylabel('这是测试的y轴名称',fontdict={'family':'simhei'})#4.2设置y轴名称
    plt.title('This is figure title')#4.3图表的标题
    plt.ylim(-1.25,1.26)
    plt.xlim(0,2)
    plt.grid(ls = "dotted",color= '#d3e3fd')
    plt.legend(loc = 1)
    #plt.savefig("./figure/second_with_style2.png",dpi= 200)
    plt.show()#3.展示图形
    
```

```python
import matplotx
plt.rcParams['font.sans-serif'] = ['Times New Roman']
with plt.style.context(matplotx.styles.ayu['light']):
    plt.figure(figsize = (10,6),dpi = 100)#创建画布
    plt.plot(x,y1,label = 'y = sin(2*np.pi*x +1) -1',)
    plt.plot(x,y2,label = 'y = cos(2*np.pi*x +1) -1',)
    plt.xlabel('这是测试的x轴名称',fontdict={'family':'simhei'})#4.1设置x轴坐标轴名称
    plt.ylabel('这是测试的y轴名称',fontdict={'family':'simhei'})#4.2设置y轴名称
    plt.title('This is figure title')#4.3图表的标题
    plt.ylim(-1.25,1.26)
    plt.xlim(0,2)
    plt.grid(ls = "dotted",color= '#d3e3fd')
    plt.legend(loc = 1)
    #plt.savefig("./figure/second_with_style3.png",dpi= 200)
    plt.show()#3.展示图形
```
