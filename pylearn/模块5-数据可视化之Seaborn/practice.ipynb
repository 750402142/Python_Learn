```python
import seaborn as sns #  pip install seaborn==0.13.2
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sys
print(f"Seaborn版本{sns.__version__}")
print('Python version:', sys.version)
print('Pandas version:', pd.__version__)
print('Numpy version:', np.__version__)
print('Matplotlib version:', matplotlib.__version__)
```

```python
tips = pd.read_csv("./data/tips.csv")
tips
```

```python
plt.rcParams['font.sans-serif'] = 'STsong'
import matplotx
with plt.style.context(matplotx.styles.pitaya_smoothie['light']):
    sns.relplot(
        data = tips,#指定数据集
        x = '小费金额 ($)',#指定x轴变量
        y = '消费金额 ($)',#指定y轴变量
        col = '就餐时间',#根据就餐时间分类,再分一个图表
        hue = '是否吸烟',#三点颜色按照是否吸烟而变化
        style = '客人性别',#三点种类按照客人性别而变化
        s = 100,#点的大小
    )
plt.savefig("./image/seaborn_01.png",dpi = 300)
```

# figure-level 级别的操作数据

```python
penguins = pd.read_csv("./data/penguins.csv")

sns.displot(
    data = penguins,
    x = '喙长 (毫米)',
    hue = '企鹅的种类',
    col = '企鹅的种类',
    palette=['cyan','red','orange'],
    binwidth = 1#线条的宽度
)
plt.savefig("./image/seaborn_02.png",dpi = 300)
```

```python
penguins.columns
```

```python
fig,axs = plt.subplots(1,2,figsize = (15,5),dpi = 100,gridspec_kw=dict(width_ratios=[4,2]))
'''subplots函数用于创建一个包含子图的图形和一组子图坐标轴对象。
1, 2表示创建一行两列的子图布局，即总共两个子图。
figsize=(15, 5)设置整个图形的大小为宽 15 单位、高 5 单位。
dpi=100设置图形的分辨率为 100 点每英寸。
gridspec_kw=dict(width_ratios=[4, 2])通过gridspec_kw参数传入一个字典，指定了子图的宽度比例为 4:2。
这意味着第一个子图的宽度是第二个子图宽度的两倍。
'''
sns.scatterplot(
    data = penguins,
    x = '喙深 (毫米)',
    y = '喙长 (毫米)',
    hue = '企鹅的种类',
    ax = axs[0]
)

sns.histplot(
    data = penguins,
    x = '企鹅的种类',
    hue = '企鹅的种类',
    shrink = 0.5,
    ax = axs[1]
)
```
