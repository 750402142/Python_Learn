```python
from shutil import posix

import pandas as pd
import numpy as np
import sys
print('Python version:', sys.version)
print('Pandas version:', pd.__version__)
print('Numpy version:', np.__version__)

from pyecharts.charts import Bar, Geo
from pyecharts.globals import JsCode
from pyecharts import options as opts
from pyecharts.globals import ThemeType, ChartType, SymbolType
```

```python
import json
with open('./data/word_data.txt','r',encoding='utf-8') as f:
    data = json.load(f)
with open('./data/crood.txt','r',encoding='utf-8') as f:
    data_loc = eval(f.read())
country = ['China','United States','Brazil','United Kingdom', 'Canda','Russia','India']
data_pair = [[x,y] for x,y in data if x in country]
data_pair.sort(key=lambda x:x[1],reverse=True)
```

```python
itemstyle = {
    'normal': {
        'borderWidth':1,
        'borderColor':JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 0, color: 'rgb(255, 191, 0)' },
          { offset: 1, color: 'rgb(224, 62, 76)' }
        ],)"""),
        'shadowColor':'green',
        'shadowBlur':1,
        'areaColor':'#2b3440',
    }
}
baritemstyle = {
    'normal': {
        'borderWidth':1,
        'color':JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 1, color: 'rgb(255, 191, 0)' },
          { offset: 0, color: 'rgb(224, 62, 76)' }
        ],)"""),
        'shadowColor':'green',
        'shadowBlur':1,
        'barBorderRadius':[10,10,10,10]
    }
}
```

```python
geo = Geo(init_opts=opts.InitOpts(theme=ThemeType.CHALK,width='100%',height='100vh'),
          is_ignore_nonexistent_coord=True)
for k,v in data_loc.items():
    geo.add_coordinate(k,v[0],v[1])
geo.add_schema(maptype='world',zoom=1,itemstyle_opts=itemstyle)
geo.add("",data_pair=data_pair,type_=ChartType.EFFECT_SCATTER)
geo.set_global_opts(title_opts=opts.TitleOpts(title = 'Bar-geo-组合图表',pos_right='center'),)
geo.set_series_opts(label_opts=opts.LabelOpts(
    is_show=True,
    position='right',
    color='white',
    font_size=16,
    font_weight='bold',
    formatter=JsCode("function(x) {console.log(x); return x.name +':'+x.value[2];}")
))
geo.render('./result/geo01.html')
```

```python
from pyecharts.charts import Bar

bar = (Bar()
    .add_xaxis([i[0] for i in data_pair])
    .add_yaxis('',[i[1] for i in data_pair],
            label_opts=opts.LabelOpts(position='insideLeft',color='black',
                formatter="{b}: {c}",font_weight='bold',font_size=14,),
            itemstyle_opts=baritemstyle
              )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='TOP  7',pos_left='5%',pos_top='55%',
         title_textstyle_opts=opts.TextStyleOpts(font_weight='bold',font_size=16)
                                 ),
        xaxis_opts=opts.AxisOpts(is_show=False),
        yaxis_opts=opts.AxisOpts(is_show=False)
    )
    .reversal_axis()
      )
bar.render('./result/new_bar01.html')
```

```python
from pyecharts.charts import Grid
grid = (Grid(init_opts=opts.InitOpts(theme =ThemeType.CHALK ,width='100%',height='100vh',))
        .add(bar,grid_opts=opts.GridOpts(pos_left='5%',pos_right='70%',pos_top='60%',))
        .add(geo,grid_opts=opts.GridOpts())
        )
grid.render('./result/组合.html')
```

```python
from pyecharts.charts import Bar,Pie
from pyecharts.faker import Faker   # 用于生成假数据
itemstyles = {
    'normal': {
        'borderRadius':15,
        'borderWidth':.5,
        'borderColor':'auto',
    }
}

va = [65,56,48,43,30,25,12]
value = [list(z) for z in zip(Faker.choose(),va)]
pie = (Pie(init_opts=opts.InitOpts(theme=ThemeType.CHALK))
    .add(series_name='系列1', data_pair=value,
         radius=['30%','80%'], # 饼图内半径和外半径
         rosetype='area',  # 否展示成南丁格尔图
         label_opts=opts.LabelOpts(formatter='{b}:{c}\n百分占比{d}%'), # 标签配置
         itemstyle_opts=itemstyles, # 图元样式配置
         emphasis_opts=opts.EmphasisOpts( is_show_label_line=True,focus='series',
                    label_opts=opts.LabelOpts(font_size=20,font_weight='bold')
                                        ), #高亮多边形配置
        )
#     .set_colors()
    .set_global_opts(tooltip_opts=opts.TooltipOpts(trigger='item'))
      )
pie.options['series'][0]['padAngle'] = 1 # 设定每个饼图各个饼块之间的距离
pie.render('./result/pie02.html')
```

```python
data = pd.read_excel('./data/gamer.xlsx')
data['Month'] = data['Month'].apply(lambda x:x.strftime('%Y-%m'))

data_pair = data[data['Name'] == 'CSGO'][['Month','Average','Peak']].values.tolist()[::-1]
from pyecharts.charts import Line
itemstyle1 = {
    'normal': {
        'color': JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 0, color: 'rgb(255, 191, 0)' },
          { offset: 1, color: 'rgb(224, 62, 76)' }
        ],)"""),
        'barBorderRadius': [20,20,20,20],      #  柱子的四个角圆角设计
        'shadowColor':'rgba(108,80,243,0.9)', # 阴影的颜色
        'shadowBlur':5, # 阴影的宽度
        'width': 4,
         'type': 'dotted'
    }
}
itemstyle2 = {
    'normal': {
        'color': JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 0, color: 'rgb(40, 141, 125)' },
          { offset: 1, color: 'rgb(199, 229, 227)' }
        ],)"""),
        'barBorderRadius': [20,20,20,20],      #  柱子的四个角圆角设计
        'shadowColor':'rgba(108,80,243,0.9)', # 阴影的颜色
        'shadowBlur':5, # 阴影的宽度
        'width': 4,
    }
}
line = (Line(init_opts=opts.InitOpts(theme=ThemeType.DARK))
    .add_xaxis(xaxis_data=[i[0] for i in data_pair])
    .add_yaxis(series_name='平均在线',y_axis=[i[1] for i in data_pair],
            is_smooth=True, # 用于平滑曲线
            is_symbol_show=False, # 关闭折线上的点
            linestyle_opts=itemstyle
               )
    .add_yaxis(series_name='在线峰值',y_axis=[i[2] for i in data_pair],
              is_smooth=True, # 用于平滑曲线
              is_symbol_show=False,
              linestyle_opts=itemstyle2,
              )
    .set_global_opts(
        datazoom_opts=opts.DataZoomOpts(type_='inside'),
        tooltip_opts=opts.TooltipOpts(trigger='axis',axis_pointer_type='cross'),
        xaxis_opts=opts.AxisOpts(name='日期',name_location='middle',name_gap=26,
                    splitline_opts={'show':False}
                                )
                    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
       )
line.render('./result/new_line.html')
```

```python
data = pd.read_excel('./data/china_province_history_data.xlsx')
dp = data[['省份','date','total_confirm']].copy()
```

```python
dp['date'] = pd.to_datetime(dp['date'])
dp.set_index('date',inplace=True)
```

```python
data_pair = dp.loc['2021-10-11'].sort_values(by = 'total_confirm').values.tolist()[-11:]
```

```python
bar = (Bar()
       .add_xaxis(xaxis_data=[i[0] for i in data_pair])
       .add_yaxis('累计人数',[i[1] for i in data_pair])
       .reversal_axis()
       )
bar.render()
```

```python
date = pd.DataFrame(
    pd.date_range('2021-01-01','2021-12-31',
                  freq='D',))[0].apply(lambda x:x.strftime('%Y-%m-%d')).tolist()
data_pairs = {}
for i in date:
    data_pairs[i] = dp.loc[i].sort_values(by = 'total_confirm').values.tolist()[-11:-1]
```

```python
data_pairs
```

```python
from pyecharts.charts import Timeline
t1 = Timeline()
for i in date:
    bar = (Bar()
           .add_xaxis(xaxis_data=[i[0] for i in data_pairs[i]])
           .add_yaxis('累计人数',[i[1] for i in data_pairs[i]])
           .reversal_axis()
           )
    t1.add(bar,i)
t1.add_schema(play_interval=300)
t1.render('./result/time_01.html')
```
