```python
from cProfile import label

import pandas as pd
import numpy as np
import sys

from pyecharts.options import InitOpts

print('Python version:', sys.version)
print('Pandas version:', pd.__version__)
print('Numpy version:', np.__version__)

from pyecharts.charts import Bar, Geo
from pyecharts.globals import JsCode
from pyecharts import options as opts
from pyecharts.globals import ThemeType, ChartType, SymbolType
```

```python
import json
with open('./data/life-expectancy-table.json','r',encoding='utf-8') as f:
    data = json.loads(f.read())
tb = pd.DataFrame(data,columns=['Income', 'Life Expectancy', 'Population', 'Country', 'Year'])
tb.drop(0,inplace=True)
tb = tb[tb['Year'] >= 1950]
countries = [
    'Finland', 'France', 'Germany', 'Iceland', 'Norway', 'Poland', 'Russia',
    'United Kingdom'
]
years = tb['Year'].unique().tolist()
```

```python
data_pairs = {}
for i in countries:
    data_pairs[i] = tb.loc[tb['Country'] == i].sort_values(by='Year')[[
        'Year', 'Income'
    ]].values.tolist()
```

```python
data_pairs
```

### 单折线

```python
from pyecharts.charts import Line

line = Line()
line.add_xaxis([str(i) for i in years])
line.add_yaxis('Russia',[i[1] for i in data_pairs['Russia']],
               is_smooth=True,
               is_symbol_show=False,
               label_opts=opts.LabelOpts(is_show=False),
               ),
line.render('./result/单折线.html')
```

### 动态折线图

```python
from pyecharts.charts import Timeline

t1 = Timeline()
for index,year in enumerate(years):
    line = Line()
    line.add_xaxis([str(i) for i in years])
    line.add_yaxis('Russia',[i[1] for i in data_pairs['Russia']][:index+1],
                   is_smooth=True,
               is_symbol_show=False,
               label_opts=opts.LabelOpts(is_show=False),
                   )
    line.set_global_opts(yaxis_opts=opts.AxisOpts(max_=70000))
    t1.add(line,year)
t1.add_schema(play_interval=400)
t1.render('./result/动态折线图.html')
```

### 动态折线图--带标记

```python
from pyecharts.charts import Timeline

t1 = Timeline()
for index,year in enumerate(years):
    markerpoints = opts.MarkPointOpts(data=[opts.MarkPointItem(
        name = 'Russia',
        coord= [str(year),data_pairs['Russia'][index][1]],
        value = data_pairs['Russia'][index][1],
        symbol_size=0)],
        label_opts=opts.LabelOpts(position = 'right',formatter="{b}:{c}",
                                  font_weight='bold',color = 'auto'),
    )
    line = Line()
    line.add_xaxis([str(i) for i in years])
    line.add_yaxis('Russia',[i[1] for i in data_pairs['Russia']][:index+1],
                   is_smooth=True,
                is_symbol_show=False,
                label_opts=opts.LabelOpts(is_show=False),
                  markpoint_opts=markerpoints,
                   )
    line.set_global_opts(yaxis_opts=opts.AxisOpts(max_=70000))
    t1.add(line,year)
t1.add_schema(play_interval=400)
t1.render('./result/动态折线图带标记.html')
```

### 动态多折线

```python
pic_dict= {
    'Russia': 'image://https://static.toolonline.net/up/2023/1229/111823090.png',
    'Finland':'image://https://static.toolonline.net/up/2023/1229/111822663.png',
    'France':'image://https://static.toolonline.net/up/2023/1229/111822034.png',
    'Germany':'image://https://static.toolonline.net/up/2023/1229/111821930.png',
    'Iceland':'image://https://static.toolonline.net/up/2023/1229/111821770.png',
    'Norway':'image://https://static.toolonline.net/up/2023/1229/111822880.png',
    'Poland':'image://https://static.toolonline.net/up/2023/1229/111822034.png',
    'Poland':'image://https://static.toolonline.net/up/2023/1229/111821984.png',
    'United Kingdom':'image://https://static.toolonline.net/up/2023/1229/111822933.png'
}
```

```python
from pyecharts.charts import Timeline

t1 = Timeline(init_opts=opts.InitOpts(theme=ThemeType.DARK,width="100%"))
def get_markerpoints(name,year,index,data_pairs):
    markerpoints = opts.MarkPointOpts(data=[opts.MarkPointItem(
        name=name,
        coord=[str(year), data_pairs[name][index][1]],
        value=data_pairs[name][index][1],
        symbol = pic_dict[name],
        symbol_size=[45,30])],
        label_opts=opts.LabelOpts(position='right', formatter="{b}:{c}",
                                  font_weight='bold', color='auto'),
    )
    return markerpoints
for index, year in enumerate(years):
    
    line = Line()
    line.add_xaxis([str(i) for i in years])
    for item in countries:
        line.add_yaxis(item, [i[1] for i in data_pairs[item]][:index + 1],
                       is_smooth=True,
                       is_symbol_show=False,
                       label_opts=opts.LabelOpts(is_show=False),
                       markpoint_opts=get_markerpoints(item,year,index,data_pairs),
                       )
    line.set_global_opts(yaxis_opts=opts.AxisOpts(max_=70000))
    t1.add(line, year)
t1.add_schema(play_interval=400,
              is_auto_play=False,
              is_loop_play=False,)
t1.options[ 'animationDurationUpdate'] = 200
t1.options['animationDuration'] = 10000
t1.options[ 'animation'] =  'auto'
t1.options[ 'animationDurationUpdate'] = 500
t1.options[ 'animationEasing'] =  'cubicInOut'
t1.options[ 'animationEasingUpdate'] = 'cubicInOut'
t1.options[ 'animationThreshold'] = 2000
t1.options[ 'progressiveThreshold'] =  3000
t1.options[ 'progressive'] =  400
t1.options[ 'hoverLayerThreshold'] =  3000
t1.render('./result/动态多折线图带标记.html')
```

```python
baritemstyle = {
    'normal': {
        'borderWidth':1,
        'color':JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 1, color: 'rgb(255, 191, 0)' },
          { offset: 0, color: 'rgb(224, 62, 76)' }
        ],)"""),
        'shadowColor':'green',
        'shadowBlur':1,
        'barBorderRadius':[10,10,10,10]
    }
}
itemstyle = {
    'normal': {
        'borderWidth':1,
        'borderColor':JsCode("""new echarts.graphic.LinearGradient(0, 0, 1, 1, [
          { offset: 0, color: 'rgb(255, 191, 0)' },
          { offset: 1, color: 'rgb(224, 62, 76)' }
        ],)"""),
        'shadowColor':'green',
        'shadowBlur':1,
        'areaColor':'#2b3440',
    }
}
```

```python
tb1 = tb.copy()
tb1.sort_values(by = ['Year','Income'],inplace=True)
data_bar = {}
for index,year in enumerate(years):
    data_bar[year] = tb1[tb1['Year']==year][['Country','Income']].values.tolist()
```

```python
t2 = Timeline()
for index, year in enumerate(years):
    bar = (Bar()
        .add_xaxis([i[0] for i in data_bar[year][-8:]])
        .add_yaxis('',[i[1] for i in data_bar[year][-8:]],
                label_opts=opts.LabelOpts(position='insideLeft',color='black',
                    formatter="{b}: {c}",font_weight='bold',font_size=14,),
                itemstyle_opts=baritemstyle
                  )
        .set_global_opts(
            title_opts=opts.TitleOpts(title='TOP  7',pos_left='5%',pos_top='55%',
             title_textstyle_opts=opts.TextStyleOpts(font_weight='bold',font_size=16)
                                     ),
            xaxis_opts=opts.AxisOpts(is_show=False),
            yaxis_opts=opts.AxisOpts(is_show=False)
        )
    .reversal_axis()
      )
    t2.add(bar,year)
t2.add_schema(play_interval=400,)
t2.render('./result/动态柱形.html')
```
