```python
import requests
import json
```

```python
names = ['太原街商圈','沈阳北站','中街商圈','棋盘山风景区','沈阳世博园']
url = 'https://api.map.baidu.com/geoconv/v2/'
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
}
ak = 'HKBiFjMeYY6TXUqUc8FRImHhMVzbwcrG'
```

```python
for i in names:
    url = f'https://map.baidu.com/?newmap=1&qt=s&da_src=searchBox.button&wd=‘+{i}+’&c=210010'
    with open(f'C:/Users/wang/Desktop/pyLearn/综合实训课程2/东软交通大数据分析/data/{i}.json', 'w', encoding='utf-8') as f:
        req = requests.get(url,headers=headers)
        f.write(json.dumps(req.json()))
    print(i,'数据获取完成')
```

```python
all_info = {}
dicts = {}
for i in names:
    with open(f'C:/Users/wang/Desktop/pyLearn/综合实训课程2/东软交通大数据分析/data/{i}.json', 'r', encoding='utf-8') as f:
        data = json.loads(f.read())
        strs = data['content'][0]['profile_geo'].split('1-')[-1].strip(';')
        croods = [','.join(i) for i in zip(strs.split(",")[::2],strs.split(",")[1::2])]
        dicts[i] = croods
        for index,item in enumerate(croods):
            params = {
                'coords': f'{item}',
                'model':'6',
                'ak':ak,
            }
            response = requests.get(url=url,headers=headers,params=params)
            print(f'正在解析:{i}中的第{index + 1}条信息,{response.json()}')
            dicts[index] = response.json()
            all_info[i] = dicts
        
```
