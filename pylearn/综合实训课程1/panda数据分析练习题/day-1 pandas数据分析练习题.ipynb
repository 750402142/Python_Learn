```python
import pandas as pd
```

```python
chipo = pd.read_csv('./data/chipotle.tsv', sep='\t')
```

```python
# 查看前十行
chipo.head(10)
```

```python
len(chipo.columns)
chipo.shape[1]#shape查看数据集形状,[行数,列数]
```

```python
chipo.groupby(by = 'item_name')['quantity'].sum().idxmax()
```

```python
#去除金额中的%$符号并转换为浮点型
chipo['item_price'].str[1:].astype('float')
chipo['item_price'].str.strip('$').astype('float')
chipo['item_price'].str.replace('$','').astype('float')
```

```python
chipo['收入'] = chipo['quantity'] * chipo['item_price']
```

```python
# 收入的平均值
chipo.groupby(by = 'order_id')['收入'].sum().mean()
print(chipo['收入'].sum()/chipo['order_id'].nunique())
```
